## Web Scraping IMDB movies

```python
import requests 
from bs4 import BeautifulSoup as bs
```

```python
r = requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")
```

```python
s = bs(r.content)
```

```python
body = s.select("tbody.lister-list tr td.titleColumn a") # Body of list of movies
movie_list = []
for link in body:
    name = link.string
    movie_list.append(name)
movie_list
```

```python
body = s.select("tbody.lister-list tr td.imdbRating strong") # Body of list of movies
rating = []
for rate in body:
    toprate = rate.string
    rating.append(toprate)
print(rating)
```

```python
print("Movie => Rating")
for i in range(250):
    print(movie_list[i]+" => "+rating[i])
with open("Movies.txt", 'a') as obj:
    for i in range(250):
        obj.write(movie_list[i]+" => "+rating[i]+" \n")
```
